```python
import numpy as np
import umap
import matplotlib.pyplot as plt

# Generating some high-dimensional random data
data = np.random.randn(1000, 10)

# Applying UMAP to reduce the dimension of the data to 2D
reducer = umap.UMAP()
embedding = reducer.fit_transform(data)

# Plot the data in the least dimensional space
plt.scatter(embedding[:, 0], embedding[:, 1])
plt.show()
```
